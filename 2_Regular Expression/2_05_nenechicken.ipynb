```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import datetime
from itertools import count
import re
import xml.etree.ElementTree as ET
import pdb
```

```python
NeNe_URL = 'https://nenechicken.com/17_new/sub_shop01.asp?page=1&ex_select=1&ex_select2=&IndexSword=&GUBUN=A'
print (NeNe_URL)
```

```python
response = urllib.request.urlopen(NeNe_URL)
```

```python
soupData = BeautifulSoup(response.read(), 'html.parser')
```

```python
print(soupData.prettify())
```

```python
store_tbls = soupData.findAll('table', attrs={'class':'shopTable'})
```

```python
print(len(store_tbls))
store_tbls[0]
```

```python
result=[]
p= re.compile(r'.+\'(.+)\'')     # map_address 에서 주소 추출 
for store_tbl in store_tbls:
    store_name = store_tbl.find('div',attrs={'class':'shopName'}).get_text()
    store_address = store_tbl.find('div',attrs={'class':'shopAdd'}).get_text().strip().split('\r')[0] # strip() : 공백 제거
    store_sido_gu = store_address.split()[:2]
    store_dong = store_address.split()[-1]    # "갈마로" 추출
    map_address = store_tbl.find('div',attrs={'class':'shopMap'}).a   # <a ....><br/>지도보기</a>
    p_address=p.match(str(map_address)).group(1)   # 강원도홍천군홍천읍갈마로4 추출
    addr_num= p_address.split(store_dong)[-1]
    #pdb.set_trace()
    
    result.append([store_name] + store_sido_gu + [store_address + ' ' + addr_num])
#    print(result)
```

# 전체 매장 주소 읽기
- 매장 페이지 수: 48 

```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import datetime
from itertools import count
import re
```

```python
result=[]
p= re.compile(r'.+\'(.+)\'')     # map_address 에서 주소 추출 

for page in range(1,47):
    NeNe_URL = 'https://nenechicken.com/17_new/sub_shop01.asp?page=%s&ex_select=1&ex_select2=&IndexSword=&GUBUN=A' % page
    response = urllib.request.urlopen(NeNe_URL)
    print("page : %s" %page)
    soupData = BeautifulSoup(response.read().decode(), 'html.parser')
    store_tbls = soupData.findAll('table', attrs={'class':'shopTable'})
    #pdb.set_trace()
    for store_tbl in store_tbls:
        store_name = store_tbl.find('div',attrs={'class':'shopName'}).get_text()
        store_address = store_tbl.find('div',attrs={'class':'shopAdd'}).get_text()
        if ( not store_address ):
            print(page)
            continue
        store_sido_gu = store_address.split()[:2]
        store_dong = store_address.split()[-1]    # "갈마로" 추출
        map_address = store_tbl.find('div',attrs={'class':'shopMap'}).a   # <a ....><br/>지도보기</a>
        p_address=p.match(str(map_address)).group(1)   # 강원도홍천군홍천읍갈마로4 추출
        addr_num= p_address.split(store_dong)[-1]
        result.append([store_name] + store_sido_gu + [store_address + ' ' + addr_num])
```
