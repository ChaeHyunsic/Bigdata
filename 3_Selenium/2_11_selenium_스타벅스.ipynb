# 서울시에 있는 스타벅스 매장 정보 크롤링

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import bs4
import time
```

```python
chromedriver_dir = r'./chromedriver.exe'
driver = webdriver.Chrome(chromedriver_dir)
```

```python
driver.get('http://www.istarbucks.co.kr/store/store_map.do')
time.sleep(5)
```

![스타벅스](../images/selenium_1.png)

![스타벅스](../images/selenium_2.png)

```python
#loca=driver.find_element_by_class_name('loca_search')
loca = driver.find_element(By.CLASS_NAME, "loca_search")
loca.click()
time.sleep(5)
```

```python
loca
```

```python
#sido=driver.find_element_by_class_name('sido_arae_box')
#driver.find_element(By.CLASS_NAME, "ico_naver").click()
sido=driver.find_element(By.CLASS_NAME, 'sido_arae_box')
li=sido.find_elements(By.TAG_NAME, 'li')
li[6].click()
time.sleep(5)
```

```python
len(li)
```

```python
#gugun=driver.find_element_by_class_name('gugun_arae_box')
#guli=gugun.find_element_by_tag_name('li')

gugun = driver.find_element(By.CLASS_NAME,'gugun_arae_box')
guli = gugun.find_element(By.TAG_NAME, "li") 
guli.click()
time.sleep(5)
```

```python
source = driver.page_source
bs = bs4.BeautifulSoup(source, 'lxml')
entire=bs.find('div', class_='loca_step3')
li_list = entire.find_all('li')
li_list
```

```python
print(entire.prettify())
```

```python
for infor in li_list:
    print(infor.find('strong').text)
```

전화번호 가져오기

```python
allText = bs.find_all('p', {'class':{'result_details'}})
for text in allText:
    print(text.get_text()[-9:])
```

특정 매장 전화번호 가져오기

```python
li_list2 = entire.find('li', {'data-name':{'울산대'}})
print(li_list2.p.text[-9:])
```

## 모든 지역의 정보 크롤링

```python
len(li)
```

```python
for i in range(0, len(li)):
    #loca=driver.find_element_by_class_name('loca_search')
    loca = driver.find_element(By.CLASS_NAME, "loca_search")
    loca.click()
    time.sleep(5)
    
    if i==16: # 세종시
        li[i].click()
        time.sleep(5)

        source = driver.page_source
        bs = bs4.BeautifulSoup(source, 'lxml')
        entire=bs.find('div', class_='loca_step3')
        li_list = entire.find_all('li')

        for infor in li_list:
            print(infor.find('p').text)
    
    else:
        li[i].click()
        time.sleep(5)
        
        #gugun=driver.find_element_by_class_name('gugun_arae_box') 
        #guli=gugun.find_element_by_tag_name('li')
        gugun = driver.find_element(By.CLASS_NAME,'gugun_arae_box')
        guli = gugun.find_element(By.TAG_NAME, "li") 
        print(guli)

        guli.click()
        time.sleep(5)

        source = driver.page_source 
        bs = bs4.BeautifulSoup(source, 'lxml')
        entire=bs.find('div', class_='loca_step3')
        li_list = entire.find_all('li')

        for infor in li_list:
            print(infor.find('p').text)
```
