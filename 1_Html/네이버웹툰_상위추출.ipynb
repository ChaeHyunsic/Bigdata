## 네이버만화 (https://comic.naver.com/index.nhn)
- 웹툰 - 장르별 추천 https://comic.naver.com/webtoon/genre.nhn
- 에피소드, 옴니버스, 스토리 별점순 상위 20개 추출

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd

genres = ['episode', 'omnibus', 'story']  # 추출해야하는 장르 list

for genre in genres:    # genres에서 순차적으로 genre를 가져옴
  # 가져온 genre에 따라 해당 장르의 url에 접근하여 html 추출
  url = 'https://comic.naver.com/webtoon/genre?order=StarScore&view=image&genre={}'.format(genre)
  html = urlopen(url)
  bs = BeautifulSoup(html, 'html.parser')

  comic_list = bs.select('ul.img_list li')    # 추출한 html중에서 원하는 정보가 담겨 있는 부분만 가져옴

  data = [['Title']]  # 추출한 data를 저장할 list
  for comic in comic_list[:20]:   # comic_list에서 상위 20개만 순차적으로 가져옴
    comic_title = comic.select_one('dl dt a')['title'].strip()    # comic에서 title 정보만 가져옴
    data.append([comic_title])  # 추출한 data 저장

  # 추출된 data를 통해 dataframe 생성
  df = pd.DataFrame(data)
  df.columns = df.iloc[0]   
  
  # 생성된 dataframe 출력
  print("{} 장르 별점순 상위 20개".format(genre))
  display(df[1:])
  print("\n\n")
```
